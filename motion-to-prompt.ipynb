```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")

inputs = tokenizer("A step by step recipe to make bolognese pasta:", return_tensors="pt")
```

```python
# Define a function to convert two rows to a tensor
def rows_to_tensor(row1, row2, row3):
    combined_row = row1 + row2 + row3
    tensor = torch.tensor([combined_row], dtype=torch.int32)
    return tensor
```

```python

import csv
# Read the CSV file
with open('output.csv', 'r') as file:
    reader = csv.reader(file)
    rows = list(reader)[1:]
```

```python
import numpy as np
tensors = []
# Combine successive rows and convert to tensors
for i in range(0, len(rows)-1, 3):
    row1 = [int(round(float(value))) for value in rows[i]]
    row2 = [int(round(float(value))) for value in rows[i+1]]
    row3 = [int(round(float(value))) for value in rows[i+2]]

    # Clip values to the range 0-32128
    row1 = np.clip(row1, 0, 24000).tolist()
    row2 = np.clip(row2, 0, 24000).tolist()
    row3 = np.clip(row3, 0, 24000).tolist()
    
    tensor = rows_to_tensor(row1, row2, row3)
    tensors.append({'input_ids': tensor, 'attention_mask': torch.ones((1,tensor.shape[1]))})
```

```python
import torch

num_tensors = 10  # Change this value to the desired number of tensors
tensors = []

for i in range(num_tensors):
    tensor = torch.randint(low=0, high=24000, size=(1, 18))
    tensors.append({'input_ids': tensor, 'attention_mask': torch.ones((1,18))})
```

```python
tensors[0]['input_ids'].shape
```

```python
prompts = [model.generate(**input) for input in tqdm(tensors)]
```

```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16)  
```

```python
pipe = pipe.to("cuda")
```

```python
for idx, prompt in enumerate(prompts):
  image = pipe(prompt).images[0]  # image here is in [PIL format](https://pillow.readthedocs.io/en/stable/)

  # Now to display an image you can either save it such as:
  #image.save(f"astronaut_rides_horse.png")

  # or if you're in a google colab you can directly display it with 
  image.save(f'generation-{idx}.png')
```
